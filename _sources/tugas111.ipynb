```python
import pandas as pd
import numpy as np
import math 
```

```python
dataset_url="https://raw.githubusercontent.com/Feb11F/datamining/main/data-penumpang-kapal-bulan-juni-2020.csv"
p=pd.read_csv(dataset_url, index_col=0)
```

```python
p
```

```python
def diss(x,y):
  a = 4
  b = 0
  for col in ["pelabuhan","kapal_berangkat","kapal_tiba","penumpang_naik","penumpang_turun"]:
    if p[col][x] == p[col][y]:
      b+=1
    return (a-b)/a
```

```python
print(diss(1,1))
```

```python
p.describe()
```

```python
#normalisasi kb
std_kb = 4.786855
mean_kb = 3.320000
for col1 in ["kapal_berangkat"]:
    p1 = p[col1]
    hasil1 = (p1-mean_kb)/std_kb
    print(hasil1)

#normalisasi kt
std_kt = 4.672436	
mean_kt = 3.366667
for col2 in ["kapal_tiba"]:
    p2 = p[col2]
    hasil2 = (p2-mean_kt)/std_kt
    print(hasil2)
```

```python
# Menghitung jarak data Numeric dengan cara Euclidean
def num_p(a,b,c):
    result = 0
    for col in [hasil1, hasil2]:
        result = math.pow(abs(col[a] - col[b]), c)
    return result**(1/c)
```

```python
print(num_p(1,2,2))
print(num_p(1,3,2))
print(num_p(1,4,2))
```

**SIMILARITY**

```python
print("Jarak d(1,2) =", diss(1,2)+num_p(1,2,2))
```

```python
print("Jarak d(1,2) =", diss(1,3)+num_p(1,3,3))
```

**Change Values to 1/0**



```python
def change_code_pekerjaan_to_biner(pekerjaan):
    return "1" if pekerjaan == "Karyawan Swasta" else "0"
```

```python
def change_code_jk_to_biner(jenis_kelamin):
    return "0" if jenis_kelamin == "perempuan" else "1"
```

```python
# Update all values series
p["pekerjaan"] = p["pekerjaan"].apply(change_code_pekerjaan_to_biner)
```

```python
# Update all values series
p["jenis_kelamin"] = p["jenis_kelamin"].apply(change_code_jk_to_biner)
```

```python
def get_series(p,idx,series):
    return p.loc[(idx),series]
```

```python
def get_dissimilarity_dataset(p,series_index = [],series = []):
    first_series = get_series(p,series_index[0],series)
    second_series = get_series(p,series_index[1],series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```
